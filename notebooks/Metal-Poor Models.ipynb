```python
def read_model_file(f):
    from astropy.io import ascii
    df=ascii.read(f).to_pandas()
    
    return df
```

```python

fig, ax=plt.subplots()
for f in fls:
    df=read_model_file(f)
    ax.plot(df['col1'], df['col2'], alpha=0.01, c='k')
    ax.set(yscale='log', xscale='log')
```
