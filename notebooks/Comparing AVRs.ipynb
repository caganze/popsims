```python
from popsims import galaxy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#compare age velocity dispersion relations

#order!
vzs= np.random.uniform(10, 21, 1000)
mask= np.argsort(vzs)
vzs=vzs[mask]
zs= np.random.uniform(0.1, 2.1, 1000 )[mask]*0.

fig, ax=plt.subplots()

ts_sharma=galaxy.avr_sharma(vzs,  z=zs, direction='vertical', verbose=False)

ax.scatter(ts_sharma[0], vzs, linewidth=2, label='Sharma', c='k', s=1)
#ax.fill_betweenx(vzs, ts_sharma[0]-ts_sharma[1], ts_sharma[0]+ts_sharma[1], color='blue')

ax.plot(galaxy.avr_aumer(vzs,  direction='vertical', verbose=False), vzs, linewidth=2, label='Aumer')

ax.plot( galaxy.avr_just(vzs,  direction='vertical', verbose=False), vzs, linewidth=2, label='Just')

ax.set(xlabel='Age (Gyr)', ylabel='Sigma Vz (km/s)')
plt.legend()
```
